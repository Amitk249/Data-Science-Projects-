Handling the outlier 

```python

# Calculate the first quartile (Q1) and third quartile (Q3)
q1 = df['Age'].quantile(0.25)
q3 = df['Age'].quantile(0.75)

# Calculate the interquartile range (IQR)
iqr = q3 - q1

# Define the lower and upper bounds for outliers
lower_bound = q1 - 1.5 * iqr
upper_bound = q3 + 1.5 * iqr

# Create a boolean mask to identify outliers
outlier_mask = (df['Age'] < lower_bound) | (df['Age'] > upper_bound)

# Remove outliers from the DataFrame
df = df[~outlier_mask]
df

```

So after Pre-processing and EDA , We have find the Correlation between 
the symtoms of the Cancer Dataset that is given, I'll pick only Four Major 
Symtoms of the cancer and compare Them with Age , Gender and their Level.

The Four Major Symtoms of Cancer are:
1) Chronic Lung Diesease
2) Passive Smoking
3) Coughing Blood
4) Clubbing of Finger Nails

```python
# lets find the correlation between them with Gender columnn 
# I'll use the corr() function  matrix 
cancer_corr = df[['Gender','chronic Lung Disease','Passive Smoker','Coughing of Blood','Clubbing of Finger Nails']].corr()
sn.heatmap(cancer_corr, annot = True)
```

There is a negative correlation (-0.203087) between gender and chronic lung disease. This means that there is a weak tendency for females to have a lower likelihood of chronic lung disease compared to males

Similarly, there is a negative correlation (-0.182674) gender and passive smoking. This indicates that females  tend to be less likely to be passive smokers compared to males

```python
# Train and Test our symtoms and then we check their accuray and importance of their features
x = major_symtoms[['chronic Lung Disease', 'Passive Smoker', 'Coughing of Blood',
       'Clubbing of Finger Nails',]]
y = major_symtoms['Gender']
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
from sklearn.linear_model import LogisticRegression
model = LogisticRegression(solver='liblinear')
model.fit(x,y)
```

```python
importances = pd.DataFrame(data={'Attribute': x.columns,'Importance': model.coef_[0]})
importances = importances.sort_values(by='Importance', ascending=False)
```

```python
LogReg = LogisticRegression(solver='liblinear')
LogReg.fit(x_train,y_train)
y_pred = LogReg.predict(x_test)
y_pred_prob = LogReg.predict_proba(x_test)
```

```python
from sklearn.metrics import (accuracy_score,classification_report,confusion_matrix)
LogRegAccuracy = accuracy_score(y_test,y_pred)
print("Accuracy : ",accuracy_score(y_test,y_pred),"\n")
print("Classification Report : \n",classification_report(y_test,y_pred))
print("Confusion Matrix : \n",confusion_matrix(y_test,y_pred),"\n")
```
