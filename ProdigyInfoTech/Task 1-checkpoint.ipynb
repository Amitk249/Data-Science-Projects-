```python

# Sample DataFrame (replace with your actual data)
data = {'Gender': ['Male', 'Female', 'Other'],
        'Population': [5000, 6000, 200]}

gender_data = pd.DataFrame(data)

# Create numerical labels for genders
gender_labels = ['Male', 'Female', 'Other']
gender_data['Gender'] = gender_data['Gender'].apply(lambda x: gender_labels.index(x))

# Create a histogram
plt.figure(figsize=(8, 6))
plt.hist(gender_data['Gender'], bins=[0, 1, 2, 3], align='mid', rwidth=0.5, color='skyblue')
plt.xlabel('Gender')
plt.ylabel('Frequency')
plt.xticks(range(len(gender_labels)), gender_labels)
plt.title('Gender Distribution (Histogram)')

# Show the chart
plt.show()
```
