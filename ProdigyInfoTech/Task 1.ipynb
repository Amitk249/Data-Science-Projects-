```python
male_data = combined_dataset[combined_dataset['WorldPopulation'] == 'male']
female_data = combined_dataset[combined_dataset['WorldPopulation'] == 'female']
years = combined_dataset.columns[1:].tolist()


male_population = male_data.iloc[:, 1:].values.flatten()
female_population = female_data.iloc[:, 1:].values.flatten()

plt.figure(figsize=(12, 6))
plt.plot(years, male_population, label='Male Population', marker='o')
plt.plot(years, female_population, label='Female Population', marker='o')
plt.xlabel('Year')
plt.ylabel('Population')
plt.title('Male and Female Population Over the Years')
plt.legend()
plt.xticks(rotation=45)
plt.grid()
plt.tight_layout()
plt.show()
```

```python

year = '2020'
data_2020 = combined_dataset[['WorldPopulation', year]]

plt.figure(figsize=(8, 6))
plt.bar(data_2020['WorldPopulation'], data_2020[year], color=['blue', 'pink'])


plt.xlabel('Gender')
plt.ylabel('Population')
plt.title(f'Population Distribution by Gender in {year}')


plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.countplot(data=combined_dataset, x='WorldPopulation', palette='pastel')


plt.xlabel('Gender')
plt.ylabel('Count')
plt.title('Gender Distribution')

plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

```python
fig, axes = plt.subplots(2, 2, figsize=(12, 10))
years = combined_dataset.columns[1:].tolist()
male_population = combined_dataset[combined_dataset['WorldPopulation'] == 'male'].iloc[:, 1:].values.flatten()
female_population = combined_dataset[combined_dataset['WorldPopulation'] == 'female'].iloc[:, 1:].values.flatten()
axes[0, 0].plot(years, male_population, label='Male Population', marker='o')
axes[0, 0].plot(years, female_population, label='Female Population', marker='o')
axes[0, 0].set_xlabel('Year')
axes[0, 0].set_ylabel('Population')
axes[0, 0].set_title('Male and Female Population Over the Years')
axes[0, 0].legend()

year = '2020'
data_2020 = combined_dataset[['WorldPopulation', year]]
sns.barplot(data=data_2020, x='WorldPopulation', y=year, palette='pastel', ax=axes[0, 1])
axes[0, 1].set_xlabel('Gender')
axes[0, 1].set_ylabel('Count')
axes[0, 1].set_title('Gender Distribution in 2020')

sns.countplot(data=combined_dataset, x='WorldPopulation', palette='pastel', ax=axes[1, 0])
axes[1, 0].set_xlabel('Gender')
axes[1, 0].set_ylabel('Count')
axes[1, 0].set_title('Overall Gender Distribution')


fig.delaxes(axes[1, 1])

plt.tight_layout()
plt.show()
```
